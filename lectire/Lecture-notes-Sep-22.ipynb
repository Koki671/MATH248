# A generic difference equation solver
What do we need to solve the following difference equation?
$$x_{n+1}=f(x_n;p)$$
where $p$ is a vector of parameters
* a function $f(x;p):\mathbb R^m\times\mathbb R^P\to\mathbb R^m$
* an initial condition $x_0$
* a parameter value $p$
* the time points $\{n:n=0,1,2,\dots, N\}$

We return a matrix, each row corresponds to a time point, each column is a variable (a component $x_i$ 

```python
from numpy import zeros

# a generic difference equation solver
# returns x0, x1, x2, ..., xN in a matrix, each row corresponds to a time point
def difference(f, x0, p, N):
    # the dimension fo the difference equations (the number of equations)
    if type(x0) is list:
        m = len(x0)
    else:
        m = 1
    # the number of times points
    # we will store the solution in V
    V = zeros((N+1, m))
    V[0] = x0
    for i in range(1, N+1):
        if p is None:
            V[i] = f(V[i-1])
        else:
            V[i] = f(V[i-1], p)
    return(V)
```

# The logistic equation
$$ X_{n+1} = rX_{n}\left(1-\frac{X_n}{K}\right)$$
Let $x_n = X_n/K$, then
$$ x_{n+1} = rx_{n}\left(1-x_n\right)$$



```python
# compute the solution to the logistic equation
# with a initial condition x0, parameter r
def logistic(x, r):
    return r*x*(1-x)

# initial condition is 0.1, parameter p(i.e., r)=1.5, return x0, ..., x50
x1 = difference(logistic, x0=0.1, p=1.5, N=50)

# initial condition is 0.8, parameter p(i.e., r)=1.5, return x0, ..., x50
x2 = difference(logistic, x0=0.8, p=1.5, N=50)

from matplotlib.pyplot import plot
plot(range(0, 51), x1, '--*r', range(0, 51), x2, '--*b')
print(x1)
```

```python
# r=0.9

x1 = difference(logistic, x0=0.1, p=0.9, N=50)
x2 = difference(logistic, x0=0.8, p=0.9, N=50)

plot(range(0, 51), x1, '--*r', range(0, 51), x2, '--*b')
```

# Periodic solution
$x_n$ is a periodic solution with period T, if $x_{n+T}=x_{n}$

```python
# r = 3.2
# 2-periodic

x1 = difference(logistic, x0=0.1, p=3.2, N=50)
x2 = difference(logistic, x0=0.8, p=3.2, N=50)

plot(range(0, 51), x1, '--*r', range(0, 51), x2, '--*b')
print(x1[46:51])
```

# Attractors
## Definition
A set $A$ is an attractor of the difference equation $x_{n+1}=f(x_n)$ if
$$A=\left\{y:\exists x_{n_k}\;s.t.\;\lim_{k\to\infty}x_{n_k}=y\right\}$$

## plot an attractor
* If we iterate the map a large number of times, hopefully $x_n$ converges
* We plot a few points after this


```python
# 3-periodic
r = 3.83
x1 = difference(logistic, x0=0.1, p=r, N=1000)

from numpy import repeat
plot(repeat(r, 100), x1[901:1001],'or')
```

# Bifurcation plot
Plot the attractor as a function of a parameter

## Example
plot the attractor of the logistic map as a function of $r$

```python
from numpy import arange
from numpy import repeat

rs = []
X = []

for r in arange(0.9, 4.3, 0.01):
    x = difference(logistic, x0=0.5, p=r, N=1000)
    # take the last 100 points
    rs = rs + repeat(r, 100).tolist()
    X = X + x[901:1001].tolist()
plot(rs, X, '.')
```

# Higher order difference equations
$$ x_{n+2} = x_{n+1} - x_{n}$$

Convert to a system of difference equations
* Let $y_n = x_{n+1}$, then
\begin{align}
x_{n+1} & = y_n\\
y_{n+1} & = x_{n+2} = y_n - x_n
\end{align}
* in vector form: $\vec X=(x,y)$
$$ f(\vec X)= \left[\begin{array}{cc}0 & 1\\-1 & 1\end{array}\right]\vec x$$

```python
from numpy import array, matmul
def f(x):
    return matmul(array([[0, 1], [-1, 1]]), x)

x = difference(f, x0=[1, -1], p=None, N=50)
plot(range(0, 51), x[:,0], '--*b')
```

```python
# the SIR model
def SIR(x, p):
    S, I, R = x
    beta, gamma, N = p
    infections = beta*S*I/N
    recoveries = gamma*I
    return [S - infections, I + infections - recoveries, R + recoveries]

# population size\
N = 1000
# initial number of infections
I0 = 10
# transmission rate
beta = 0.5
# recovery rate
gamma = 0.25
x = difference(SIR, x0=[N-I0, I0, 0], p=[beta, gamma, N], N=50)
plot(range(0, 51), x[:,1], '--*b')
```
